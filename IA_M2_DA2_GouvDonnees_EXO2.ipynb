## **(a) Chiffrer le jeu de données en utilisant des bibliothèques Python comme cryptography ou PyCrypto**

```python
# Generation d'une clé pour le cryptage et le décryptage
key = Fernet.generate_key()

#Création d'un objet qui permet de chiffrer et déchiffrer les données en utilisant la clé fournie
cipher_suite = Fernet(key)
```

```python
# Affichage de la clé
key
```

```python
# Creation d'une fonction d'encryptage
def encrypt_data(data):
    return cipher_suite.encrypt(data.encode()).decode()
```

```python
# Application de la fonction aux différentes colonnes
data['first_name_encrypted'] = data['first_name'].apply(encrypt_data)
data['last_name_encrypted'] = data['last_name'].apply(encrypt_data)
data['email_encrypted'] = data['email'].apply(encrypt_data)
data['ip_address_encrypted'] = data['ip_address'].apply(encrypt_data)
```

```python
# Supprimer les colonnes initiales pour ne garder que celles cryptées
encrypted_data = data.drop(['first_name', 'last_name', 'email', 'ip_address'], axis=1)
```

La mise en place d’un contrôle d’accès basé sur les rôles (RBAC) consiste à définir des règles qui régissent l’accès aux données sensibles en fonction du rôle d’un utilisateur. Voici comment tu peux procéder pour implémenter un tel système dans ton jeu de données.

Étapes pour implémenter RBAC
Définir les rôles : Chaque utilisateur se voit attribuer un rôle qui détermine les champs auxquels il peut accéder.
Attribuer des permissions aux rôles : Chaque rôle dispose de permissions spécifiques qui lui permettent ou non d'accéder à certains champs (par exemple, les champs sensibles comme les emails ou les numéros de téléphone).
Appliquer le filtrage des données : En fonction du rôle de l'utilisateur, certaines données seront masquées ou accessibles.

```python
# Définir les rôles et les permissions associées
roles_permissions = {
    'admin': ['id', 'first_name', 'last_name', 'email', 'gender', 'ip_address'],  # Accès complet
    'manager': ['id', 'first_name', 'last_name', 'gender'],  # Accès restreint, pas d'accès à l'email et à l'ip
    'guest': ['id', 'gender']  # Très restreint, seulement accès aux identifiants non sensibles
}
```

```python
# Attribuer un rôle (simulons que cet utilisateur soit un 'guest')
user_role = 'manager'  # Par exemple, cela pourrait être dynamique selon l'utilisateur connecté
```

```python
# Obtenir les permissions de l'utilisateur en fonction de son rôle
allowed_fields = roles_permissions[user_role]
```

## Politique de Protection des Données : Chiffrement et Contrôles d'Accès Basés sur les Rôles (RBAC)

**1. Introduction**

Cette politique a pour objectif de décrire les mesures de sécurité mises en place pour protéger la confidentialité des données sensibles au sein de l'organisation. Elle spécifie l'utilisation du chiffrement pour protéger les données en transit et au repos, ainsi que la mise en œuvre de contrôles d'accès basés sur les rôles (RBAC) pour restreindre l'accès aux données sensibles en fonction des responsabilités des utilisateurs.


**2. Objectifs**

Les objectifs de cette politique sont :

Garantir la confidentialité des données sensibles.
Protéger les données contre l'accès non autorisé en mettant en place des mécanismes de chiffrement et de contrôles d'accès.
Assurer que seuls les utilisateurs ayant les autorisations appropriées peuvent accéder à certaines informations sensibles (par exemple, les emails, les numéros de téléphone, etc.).


**3. Champ d'application**

Cette politique s'applique à toutes les données sensibles de l'organisation, notamment :

Données personnelles (prénoms, noms de famille, emails, numéros de téléphone).
Données financières et professionnelles sensibles.
Toute autre donnée jugée confidentielle par la direction de l'organisation.


**4. Chiffrement des données**


  **4.1. Chiffrement des données au repos**

  
Toutes les données sensibles stockées dans les bases de données doivent être chiffrées afin de garantir leur confidentialité. Le chiffrement utilisé est symétrique, basé sur la bibliothèque cryptography de Python, et respecte les normes de chiffrement AES-256.

Clé de chiffrement : La clé de chiffrement est générée dynamiquement et doit être sauvegardée dans un emplacement sécurisé, tel qu'un gestionnaire de clés (par exemple, AWS KMS ou Azure Key Vault).

Processus de chiffrement : Les données sensibles telles que les prénoms, noms de famille, emails et adresses IP sont chiffrées avant d'être stockées dans la base de données.
Exemples de données chiffrées :
Prénoms, Noms de famille.
Emails.
Adresses IP.

**4.2. Chiffrement des données en transit**

Les communications impliquant des données sensibles (par exemple, entre le serveur et l'utilisateur) doivent être sécurisées à l'aide du protocole HTTPS/TLS pour garantir la protection des données en transit.

**5. Contrôles d'Accès Basés sur les Rôles (RBAC)**

**5.1. Définition des rôles**

L'accès aux données sensibles est restreint en fonction du rôle de chaque utilisateur au sein de l'organisation. Les rôles doivent être clairement définis et attribués en fonction des responsabilités.

Les rôles sont définis comme suit :

Admin : Accès complet à toutes les données, y compris les données sensibles.
Manager : Accès aux informations de base, mais aucune autorisation pour consulter les emails, adresses IP ou autres informations sensibles.
Guest : Accès restreint uniquement aux données non sensibles telles que l'identifiant et le sexe.

**5.2. Gestion des permissions**

Chaque rôle dispose de permissions spécifiques lui permettant d'accéder à certaines colonnes du jeu de données.

Admin : Peut consulter toutes les colonnes, y compris les champs sensibles comme les emails et les adresses IP.
Manager : Accès aux colonnes non sensibles comme le prénom, nom, mais sans accès aux emails et adresses IP.
Guest : Ne peut consulter que les colonnes non sensibles telles que l'identifiant et le genre.

**5.3. Mise en œuvre**

Le contrôle d'accès basé sur les rôles est implémenté en appliquant un filtrage dynamique sur les colonnes du jeu de données en fonction du rôle de l'utilisateur. Ce filtrage est appliqué avant l'affichage des données à l'utilisateur.

**6. Sauvegarde et récupération des clés de chiffrement**

Les clés de chiffrement doivent être sauvegardées dans un gestionnaire de clés sécurisé et récupérées uniquement par les utilisateurs ou services autorisés. En cas de perte de la clé, un protocole de récupération de clé doit être mis en place pour minimiser l'impact sur les données chiffrées.

**7. Responsabilités**

Administrateurs système : Doivent garantir que les données sensibles sont toujours chiffrées et que les clés sont gérées de manière sécurisée.
Développeurs : Sont responsables de l'intégration du chiffrement et du contrôle d'accès dans les systèmes et les applications.
Responsables de la conformité : Doivent veiller à ce que les processus de chiffrement et d'accès respectent les réglementations en vigueur (RGPD, HIPAA, etc.).

**8. Formation et sensibilisation**

Tous les employés doivent recevoir une formation régulière sur l'importance de la confidentialité des données et la manière dont le chiffrement et les contrôles d'accès basés sur les rôles (RBAC) sont mis en œuvre au sein de l'organisation.

**9. Audit et conformité**

Des audits réguliers doivent être effectués pour vérifier que les politiques de chiffrement et de contrôle d'accès sont bien mises en œuvre et respectées. Toute déviation doit être corrigée immédiatement.

**10. Révision de la politique**

Cette politique doit être révisée au moins une fois par an pour s'assurer qu'elle reste conforme aux réglementations en vigueur et qu'elle reflète les meilleures pratiques en matière de protection des données.
